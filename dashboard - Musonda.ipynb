## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# Create a Title for the Dashboard
title = "Real Estate Analysis of Toronto from 2001 - 2016"
welcome_tab = pn.Row(pn.Column(title), neighbourhood_map())

# Define a welcome text
welcome_text = 'Here is the real estate analysis of Toronto from 2001 - 2016. You will find information related to costs trends over the years, neighbourhood anaylsis, and information about the top ten most expensive neighbourhoods of Toronto. Data used from years 2001, 2006, 2011, and 2016. Enjoy!'


# Create the main dashboard

welcome_column = pn.Column(
    welcome_text,
    fig)

year_analysis_column = pn.Column(
    create_bar_chart(sliced_data_2001, title ='Dwelling Types in Toronto in 2001', xlabel='2001', ylabel='Dwelling Type Units', color='red'),
    create_bar_chart(sliced_data_2006, title = "Dwelling Types in Toronto in 2006",xlabel= "2006",ylabel= "Dwelling Type Units",color= "blue"),
    create_bar_chart(sliced_data_2011, "Dwelling Types in Toronto in 2006", "2006", "Dwelling Type Units", "yellow"),
    create_bar_chart(sliced_data_2016,title ='Dwelling Types in Toronto in 2016', xlabel='2016', ylabel='Dwelling Type Units', color='pink')
)

shelter_costs_vs_house_value_column = pn.Column(
    average_monthly_costs['shelter_costs_owned'].hvplot(title = 'Average Monthly Shelter Cost for Owned Dwellings in Toronto', xlabel = 'Year', ylabel = 'Avg Monthly Shelter Costs', color = 'blue'),
    average_monthly_costs['shelter_costs_rented'].hvplot(title = 'Average Monthly Shelter Cost for Rented Dwellings in Toronto', xlabel = 'Year', ylabel = 'Avg Monthly Shelter Costs', color = 'yellow'),
    average_house_value.hvplot(title = 'Average House Value in Toronto', xlabel = 'Year', ylabel = 'Avg House Value', color = 'orange')
    )

neighbourhood_analysis_column = pn.Column(
    interact(choose_neighbourhood_value, neighbourhood=neighbourhood_data["neighbourhood"],width=600,height=500),
    interact(choose_neighbourhood, neighbourhood=all_dwelling_types_df["neighbourhood"], width=600, height=500)
    )

expensive_neighbourhood_column = pn.Column(
    expensive_neighbourhoods.hvplot.bar(title="Top 10 Expensive Neighborhoods in Toronto", xlabel="Neighbourhood",ylabel="Average House Value", rot=90,color ='green')
    )

# Create a tab layout for the dashboard
dashboard = pn.Tabs(
    ("Welcome",welcome_column),
    ("Year Analysis",year_analysis_column),
    ("Shelter Costs vs House Value",shelter_costs_vs_house_value_column), 
    ("Neighbourhood Analysis",neighbourhood_analysis_column), 
    ("Top Ten Expensive Neighbourhoods",expensive_neighbourhood_column)
    )
```
